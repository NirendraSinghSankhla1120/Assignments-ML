# Step:-1

## Importing all the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder,StandardScaler,MinMaxScaler
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

## Missing Values

```python

missing=['?','-']
df_data=pd.read_csv('adultIn.csv',na_values=missing)
df_data
```

```python
df_data.isnull().sum()
```

```python
df_data.dropna(axis=0,inplace=True)
```

# Step:-2 

## Data Preparation

```python
# droping the column
df_data.drop(['fnlwgt'],axis=1,inplace=True)
```

```python
#extracting columns consisting numeric values
df_num=df_data.select_dtypes(include='number')
df_num.head()
```

```python
#extract columns consisting categorial values
df_cat=df_data.select_dtypes(include='object')
df_cat.head()
```

```python
# Standarizing numerical data
scale=StandardScaler()
df_ss=pd.DataFrame(scale.fit_transform(df_num),columns=df_num.columns)
df_ss.head()
```

```python
# encoding categorical features
ohe = OneHotEncoder(drop='first', sparse=False)
df_ohe=pd.DataFrame(ohe.fit_transform(df_cat),columns=ohe.get_feature_names(df_cat.columns))
df_ohe.head()
```

```python
#concatinating both numeric and categorial columns
df=pd.concat([df_ss,df_ohe],axis=1)
pd.set_option('display.max_rows',100)
df
```

```python
# setting features and labels
y=df.pop('income_>50K')
x=df
```

```python
#Train-test Split
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(x, y, train_size=0.2)
```

# Step:-3,4 and 5

## 1-Traning the model
## 2-Predict on test data
## 3-Evaluate model

```python
# using logistic regression bcoz of classification problem
lr=LogisticRegression(max_iter=500)
lr.fit(xtrain,ytrain)
y_pred=lr.predict(xtest)
metrics.accuracy_score(ytest,y_pred)
```

```python
confusion=metrics.confusion_matrix(ytest,y_pred)
confusion
```

```python
print(metrics.classification_report(ytest,y_pred))
```

# Step:-6

## RFE

```python
from sklearn.feature_selection import RFE
rfe=RFE(lr,5)
rfe=rfe.fit(xtrain,ytrain)
```

```python
temp_df=pd.DataFrame({'Columns':xtrain.columns,'Included':rfe.support_,'Ranking':rfe.ranking_})
temp_df
```

```python
imp_column=xtrain.columns[rfe.support_]
imp_column
```

```python
x_train_rfe=xtrain[imp_column]
x_train_rfe.head()
```

```python
lr_rfe=LogisticRegression(max_iter=500)
lr_rfe.fit(x_train_rfe,ytrain)
```

```python
x_test_rfe=xtest[imp_column]
y_pred_rfe=lr_rfe.predict(x_test_rfe)
```

```python
metrics.accuracy_score(ytest,y_pred_rfe)
```

# Step:-7

## PCA
